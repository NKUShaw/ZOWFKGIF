```python
import matplotlib.pyplot as plt
```

```python
# 
remove_entities = [500, 1000, 1500, 2000, 2500, 3000]
retrain_hit1 = [0.180, 0.165, 0.155, 0.157, 0.144, 0.135]
gif_hit1 = [0.166, 0.150, 0.143, 0.135, 0.126, 0.123]
baseline_hit1 = 0.194

# 
plt.figure(figsize=(10, 5))
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

# 
plt.plot(remove_entities, retrain_hit1, label='Retrain', marker='o')
plt.plot(remove_entities, gif_hit1, label='GIF', marker='o')

# 
plt.axhline(y=baseline_hit1, color='r', linestyle='--', label='Full dataset trained hit@1')

# 
plt.title('Hit@1 vs. Removed Entities')
plt.xlabel('Removed Entities')
plt.ylabel('Hit@1')

# 
plt.ylim(0, 0.2)

# 
plt.legend()

# 
plt.grid(True)
plt.savefig('pic_Hit@1.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
